         │
         ▼
┌─────────────────┐
│  Bytecode       │  ← Intermediate form (.pyc files)
└────────┬────────┘
         │
         ▼
┌─────────────────┐
│  Python Virtual │
│  Machine (PVM)  │  ← Executes bytecode instructions
└────────┬────────┘
         │
         ▼
┌─────────────────┐
│  Computer       │
│  executes!      │  ← Actions happen!
└─────────────────┘

### 2. Float (`float`) - Decimal Numbers

**Floats** are numbers with **decimal points**.

Examples: `3.14`, `0.5`, `2.0`, `-1.75`, `0.001`

Think of measuring things precisely: height (1.75 meters), weight (68.5 kg), temperature (36.6°C)

**The key difference from integers**: THE DECIMAL POINT!

<div dir="rtl">

### 2. Float (`float`) - أرقام عشرية

</div>

<div dir="rtl">

الـ **Float** هي أرقام **بفاصلة عشرية**.

</div>

<div dir="rtl">

أمثلة: `3.14`, `0.5`, `2.0`, `-1.75`, `0.001`

</div>

<div dir="rtl">

تخيل قياس الأشياء بدقة: الطول (1.75 متر)، الوزن (68.5 كيلو)، درجة الحرارة (36.6 درجة)

</div>

<div dir="rtl">

**الفرق الأساسي عن الـ integer**: الفاصلة العشرية!

</div>

Let's see floats in action:


````python
# Different floats
pi = 3.14159
half = 0.5
tiny_number = 0.001
two_as_float = 2.0  # Has a decimal point, so it's a float!

print("Pi:", pi)
print("Half:", half)
print("Tiny:", tiny_number)
print("Two (float):", two_as_float)
````

````python
# Compare: Integer 2 vs Float 2.0
integer_two = 2
float_two = 2.0

print("Integer 2:")
print("  Value:", integer_two)
print("  Type:", type(integer_two))

print("\nFloat 2.0:")
print("  Value:", float_two)
print("  Type:", type(float_two))

print('\nThey look the same, but Python treats them differently!')
````

````python
# Math with floats
height = 1.75  # meters
weight = 68.5  # kg

print("Height:", height, "m")
print("Weight:", weight, "kg")
print("Height + Weight:", height + weight)
print("Height × 2:", height * 2)
````

````python
# Floats are VERY important in machine learning!
# Here are some common ML values:

learning_rate = 0.01    # How fast a model learns (small decimal)
accuracy = 0.95         # Model accuracy: 95% = 0.95
loss = 0.234            # Model error

print("Learning rate:", learning_rate)
print("Accuracy:", accuracy)
print("Loss:", loss)

print("\nNotice: All these ML values are floats!")
````

**Key Takeaway**:

- **Integers** = whole numbers: `1, 2, 3, 100`
- **Floats** = decimal numbers: `1.5, 2.0, 0.001, 3.14159`

In machine learning, we use floats A LOT because we need precise measurements!

**Important**: `True` and `False` are special words in Python (notice the capital letters!)

<div dir="rtl">

### 4. Boolean (`bool`) - صح أو خطأ

<div dir="rtl">

<div dir="rtl">

Let's see booleans:


````python
# Simple booleans
is_sunny = True
is_raining = False

print("Is it sunny?", is_sunny)
print("Is it raining?", is_raining)
print("Type:", type(is_sunny))
````

````python
# Booleans come from comparisons
age = 25

is_adult = age >= 18
is_child = age < 13
is_exactly_25 = age == 25  # Note: == for comparison, = for assignment

print("Age:", age)
print("Is adult (≥18)?", is_adult)
print("Is child (<13)?", is_child)
print("Is exactly 25?", is_exactly_25)
````

````python
# Comparison operators
a = 10
b = 20

print("")
print("a == b (equal):", a == b)
print("a != b (not equal):", a != b)
print("a < b (less than):", a < b)
print("a > b (greater than):", a > b)
print("a <= b (less or equal):", a <= b)
print("a >= b (greater or equal):", a >= b)
````

````python
# Booleans in machine learning
is_trained = False        # Has the model finished training?
use_gpu = True            # Should we use GPU for faster computation?
model_converged = False   # Did training reach optimal point?

print("Model trained?", is_trained)
print("Use GPU?", use_gpu)
print("Converged?", model_converged)

print("\nBooleans help us make decisions in our code!")
````

**Key Takeaway**:

- Booleans = `True` or `False` only
- Come from comparisons (`<`, `>`, `==`, etc.)
- Used for decisions and conditions

### Type Conversion: Changing Between Types

Sometimes we need to convert data from one type to another. Let's see how:

<div dir="rtl">

### تحويل الأنواع: التغيير بين الأنواع

</div>

<div dir="rtl">

أحياناً نحتاج تحويل البيانات من نوع لآخر. تعالى نشوف
:

</div>


````python
# String to Integer .. casting
age_text = "25"
print(f"Original: '{age_text}' (type: {type(age_text).__name__})")

age_number = int(age_text)
print(f"Converted: {age_number} (type: {type(age_number).__name__})")

# Now we can do math!
print(f"Next year: {age_number + 1}")
````

````python
# String to Float
price_text = "19.99"
print(f"Original: '{price_text}' (type: {type(price_text).__name__})")

price_number = float(price_text)
print(f"Converted: {price_number} (type: {type(price_number).__name__})")

print(f"With tax: ${price_number * 1.1:.2f}")
````

````python
# Number to String
count = 42
print(f"Original: {count} (type: {type(count).__name__})")

count_text = str(count)
print(f"Converted: '{count_text}' (type: {type(count_text).__name__})")

# Now we can combine it with other text
message = "You have " + count_text + " items"
print(message)
````

````python
# Float to Integer (loses decimal part!)
pi = 3.14159
print(f"Original: {pi}")

pi_as_int = int(pi)
print(f"As integer: {pi_as_int}")
print("⚠️ Warning: Decimal part was lost!")
````

````python
# Creating lists
fruits = ["apple", "banana", "cherry"] # square brackets
numbers = [10, 20, 30, 40, 50]
mixed = ["hello", 42, 3.14, True]  # Can mix types!

print("Fruits:", fruits)
print("Numbers:", numbers)
print("Mixed:", mixed)
````

````python
# Empty list
empty_list = ["hello", 42, 3.14, True]
print("Empty list:", empty_list)
print("Length:", len(empty_list)) # length
````

### Accessing List Elements (Indexing)

Use square brackets `[]` with the position number (index):

<div dir="rtl">

### الوصول لعناصر الـ List (Indexing)

</div>

<div dir="rtl">

استخدم الأقواس المربعة `[]` مع رقم الموقع (index):

</div>

````python
# ML Example: Accuracy function
def calculate_accuracy(predictions, true_labels):
    """
    Calculate classification accuracy.

    Args:
        predictions: List of predicted labels
        true_labels: List of true labels

    Returns:
        Accuracy as percentage (0-100)
    """
    correct = 0
    for i in range(len(predictions)):
        if predictions[i] == true_labels[i]:
            correct += 1  # this is the same as correct = correct + 1

    accuracy = (correct / len(predictions)) * 100
    return accuracy


# Test it
pred = [0, 1, 1, 0, 1]
true = [0, 1, 0, 0, 1]
acc = calculate_accuracy(pred, true)
print(f"Accuracy: {acc:.1f}%")
````

````python
# Euclidean distance
def distance(point1, point2):
    """
    Calculate distance between two 2D points.

    Formula: sqrt((x1-x2)² + (y1-y2)²)
    """
    x1, y1 = point1
    x2, y2 = point2

    dist = ((x1-x2)**2 + (y1-y2)**2)**0.5
    return dist


# Test
p1 = [0, 0]
p2 = [3, 4]
d = distance(p1, p2)
print(f"Distance from {p1} to {p2}: {d}")
````

````python
# Normalization function (preview this will be discussed in the lecture in detail but see how to implement it here!)
def normalize(values):
    """
    Normalize values to [0, 1] range.

    Formula: (x - min) / (max - min)
    """
    min_val = min(values)
    max_val = max(values)
    range_val = max_val - min_val

    normalized = []
    for val in values:
        norm_val = (val - min_val) / range_val
        normalized.append(norm_val)

    return normalized


# Test
data = [10, 20, 30, 40, 50]
norm_data = normalize(data)
print(f"Original: {data}")
print(f"Normalized: {[f'{x:.2f}' for x in norm_data]}")
````

````python
# Given data
data_points = [1.5, 4.2, 2.8, 5.6, 3.1]
true_labels = [0, 1, 0, 1, 1]

# Task 1: Make predictions (if > 3.0, predict 1, else 0)
# YOUR CODE HERE
def make_predictions(data_points , true_lables):
  predictions =[]
  for point in data_points :
    if ( point > 3.0 ):
      predictions.append(1)
    else :
      predictions.append(0)
  return predictions
predictions = make_predictions(data_points , true_labels)


# Task 2: Calculate accuracy
# YOUR CODE HERE
def calc_accuracy (predictions , true_lables):
  correct = 0.0
  for idx in range (len(predictions)):
    if (predictions[idx] == true_lables[idx]):
      correct +=1
  return (correct/len(predictions))*100
accuracy = calc_accuracy(predictions , true_labels)

# Task 3: Normalize data
# YOUR CODE HERE
def normalize_list (data_points):
  min_val = min(data_points)
  max_val = max(data_points)
  range = max_val - min_val
  normalized_list = []
  for point in data_points :
    norm_val = (point - min_val)/range
    normalized_list.append(norm_val)
  return normalized_list
normalized_data = normalize_list(data_points)

# Results
print("╔═══════════════════════════════╗")
print("║   ML Pipeline Results        ║")
print("╚═══════════════════════════════╝")
print(f"\nData: {data_points}")
print(f"Normalized: {[f'{x:.3f}' for x in normalized_data]}")
print(f"\nTrue: {true_labels}")
print(f"Pred: {predictions}")
print(f"\nAccuracy: {accuracy:.1f}%")
````
